```python
import pandas as pd
praha = pd.read_csv("zam_praha.csv")
plzen = pd.read_csv("zam_plzeň.csv")
liberec = pd.read_csv("zam_liberec.csv")
praha['mesto'] = ["Praha"] * len(praha)
plzen['mesto'] = ["Plzen"] * len(plzen)
liberec['mesto'] = ["Liberec"] * len(liberec)

zamestnanci = pd.concat([praha, plzen, liberec], ignore_index=True)
zamestnanci.head()
```

```python
platy = pd.read_csv("platy_2021_02.csv")
zamestnanci_platy = pd.merge(zamestnanci, platy, how="inner", on=["cislo_zamestnance"])
zamestnanci_platy.head(20)
```

```python
print(zamestnanci.shape)
print(zamestnanci_platy.shape)
```

```python
zamestnanci_platy.groupby("mesto")["plat"].mean()
```

```python
vykazy = pd.read_csv("vykazy.csv")

zamestnanci_vykazy = pd.merge(vykazy, zamestnanci, how='left', left_on='emloyee_id', right_on='cislo_zamestnance')
zamestnanci_vykazy.head()

```

```python
zamestnanci_vykazy.groupby("project")["hours"].sum()
```
